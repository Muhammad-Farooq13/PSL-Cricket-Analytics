# PSL Dataset - Exploratory Data Analysis

This notebook performs exploratory data analysis on the PSL Complete Dataset (2016-2024).

## Objectives:
1. Load and inspect the dataset
2. Understand data structure and types
3. Identify missing values and outliers
4. Visualize key patterns and relationships
5. Generate insights for feature engineering

```python
# Import required libraries
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.load_data import DataLoader
from src.visualization.visualize import DataVisualizer

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries imported successfully!")
```

## 1. Load Dataset

```python
# Initialize data loader
loader = DataLoader(data_path='../data/raw')

# Load PSL dataset
df = loader.load_psl_dataset()

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

## 2. Initial Data Inspection

```python
# Display first few rows
print("First 5 rows of the dataset:")
df.head()
```

```python
# Display data types and non-null counts
print("Dataset Info:")
df.info()
```

```python
# Display summary statistics
print("Summary Statistics:")
df.describe()
```

## 3. Missing Values Analysis

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Percentage': missing_percent
}).sort_values('Missing_Count', ascending=False)

print("Missing Values Summary:")
print(missing_df[missing_df['Missing_Count'] > 0])
```

```python
# Visualize missing values
if missing_df['Missing_Count'].sum() > 0:
    plt.figure(figsize=(12, 6))
    missing_data = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count')
    plt.barh(missing_data.index, missing_data['Percentage'])
    plt.xlabel('Percentage Missing')
    plt.title('Missing Values by Column')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in the dataset!")
```

## 4. Duplicate Analysis

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
print(f"Percentage of duplicates: {(duplicates / len(df)) * 100:.2f}%")
```

## 5. Data Distribution Analysis

```python
# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"Numerical columns ({len(numerical_cols)}): {numerical_cols}")
print(f"\nCategorical columns ({len(categorical_cols)}): {categorical_cols}")
```

```python
# Visualize numerical distributions (first 6 columns)
if len(numerical_cols) > 0:
    visualizer = DataVisualizer(output_dir='../reports/figures')
    cols_to_plot = numerical_cols[:6]  # Plot first 6 numerical columns
    visualizer.plot_distribution(df, cols_to_plot, save_name='numerical_distributions.png')
```

## 6. Correlation Analysis

```python
# Correlation matrix for numerical features
if len(numerical_cols) > 1:
    visualizer.plot_correlation_matrix(df[numerical_cols], save_name='correlation_matrix.png')
```

## 7. Categorical Analysis

```python
# Analyze categorical columns
if len(categorical_cols) > 0:
    for col in categorical_cols[:5]:  # Analyze first 5 categorical columns
        print(f"\nUnique values in {col}: {df[col].nunique()}")
        print(df[col].value_counts().head(10))
```
